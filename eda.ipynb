```python
x = pd.read_csv("./IPLMatches.csv")
```

```python
x
```

```python
#Toss winners 
x['player_of_match'].value_counts()
```

```python
players=list(x['player_of_match'].unique())
len(players)
```

```python
#Making a new Dataset which is required for the bar chart race
data=np.zeros((816,len(players)))
new_x=pd.DataFrame(data,index=x['date'],columns=players)
new_x
```

```python
# Sample Dictionary which we will use in the new Dataset.
players_won={}
for player in players:
    players_won[player]=0
players_won
```

```python
# Creation of the new Dataset.
for match in tqdm(range(0,len(x))):
    motm=x['player_of_match'].iloc[match]
    for player in players_won:
        if motm==player:
            players_won[player]=players_won[player]+int(1)
            new_x[player].iloc[match]=players_won[player]
        else:
            new_x[player].iloc[match]=players_won[player]
```

```python
new_x
```

```python
#Deleting rows with the same date

new_x=new_x.reset_index()
delete_index=[]
for match in tqdm(range(0,len(new_x))):
    try:
        if new_x['date'].iloc[match]==new_x['date'].iloc[match+1]:
            delete_index.append(match)
    except:
        pass

new_x = new_x.drop(new_x.index[delete_index])
```

```python
final_x=new_x.set_index('date')
```
